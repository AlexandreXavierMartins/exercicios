```python
import math
import matplotlib.pyplot

def f(x):
  return x**2 - math.cos(4*math.pi*x)

#print(f(0))
#print(f(10))
a = -3
b = 3
e = 0.001
cont = 0
x = a
f_star = f(x)
x_star = x

labelx = []
labely = []
while x <= b:
  labelx.append(x)
  labely.append(f(x))
  x = x + e
  cont = cont +1
  if f(x) < f_star:
    f_star = f(x)
    x_star = x 
print('A melhor solução é', format(x_star, '.2f'), 'de custo', format(f_star, '.2f'))
print('Número de iterações', cont)
matplotlib.pyplot.plot(labelx, labely)
matplotlib.pyplot.show()
```
